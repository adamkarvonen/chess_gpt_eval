```python
import numpy as np
import matplotlib.pyplot as plt

def get_player_two_skill(base=1.3, uniform_fraction=0.1):
    # 10% of the time, choose a random skill between -2 and 20
    if np.random.random() < uniform_fraction:
        return np.random.randint(-2, 21)

    # 90% of the time, use exponential distribution
    else:
        skill_levels = np.arange(-2, 21)
        weights = base ** skill_levels
        normalized_weights = weights / sum(weights)
        return np.random.choice(skill_levels, p=normalized_weights)

def visualize_distribution(samples, bins=23):
    plt.hist(samples, bins=bins, range=(-2.5, 20.5), density=True, alpha=0.7)
    plt.xlabel("Skill Level")
    plt.ylabel("Probability Density")
    plt.title("Distribution of Player Two Skill Level")
    plt.xticks(np.arange(-2, 21))
    plt.grid(axis='y')
    plt.show()

if __name__ == "__main__":
    # Sample skill levels for visualization
    samples = [get_player_two_skill() for _ in range(10000)]
    visualize_distribution(samples)
```
