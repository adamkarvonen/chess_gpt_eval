```python
import pandas as pd
import matplotlib.pyplot as plt

input_file = 'lichess_9gb.csv'
output_file = input_file.replace('.csv', '_200k_elo_bin.csv')
# Sample DataFrame
graph_df = pd.read_csv(input_file)

# Define bins
bins = range(500, 3500, 100)

# Bin the data
graph_df['elo_bin'] = pd.cut(graph_df['WhiteElo'], bins=bins, right=False)

# Get the games count in each bin
bin_counts = graph_df['elo_bin'].value_counts()

# Create a new DataFrame for the limited data
limited_data = pd.DataFrame()

for bin_range, count in bin_counts.items():
    # Limit to 200,000 games per bin, or take all if less than 200,000
    limited_count = min(count, 200000)
    # Get a sample of games from each bin
    limited_data = pd.concat([limited_data, graph_df[graph_df['elo_bin'] == bin_range].sample(n=limited_count)])

limited_data.to_csv(output_file)

print(len(limited_data))
print(len(graph_df))

# Plot the histogram using the limited data
plt.hist(limited_data['WhiteElo'], bins=bins, edgecolor='black')
plt.title('Distribution of White Player Elo Ratings (Limited to 200k per Bin)')
plt.xlabel('Elo Rating')
plt.ylabel('Count')
plt.show()
```
